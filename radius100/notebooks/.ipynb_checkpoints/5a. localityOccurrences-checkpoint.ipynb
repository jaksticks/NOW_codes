# WEIGHTED OCCURRENCES IN FOCAL AREAS

Here we look for each locality, which other localities exist in a given area around the focal locality in the current timeunit, in the preceding time unit and in the next timeunit. For each focal locality, we calculate the number of occurrences in the focal area for each of the time units and we weight these occurrences based on their distance from the center of the focal area.

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
import time
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df = pd.read_csv('../csv/NOW_dataframe_cleaned.csv', index_col=0)
localityFeatures = pd.read_csv('../csv/localityFeatures.csv', index_col=0)
timeUnits = ['old','MN1','MN2','MN3','MN4','MN5','MN6','MN7-8','MN9','MN10','MN11','MN12','MN13','MN14','MN15','MN16','MN17','MQ18','MQ19','recent']

pd.set_option('max_columns',100)
```

```python
# USING ONLY LOCALITIES IN EUROPE AND CLOSE BY
localityFeatures = localityFeatures[(localityFeatures['LAT']>35) & (localityFeatures['LONG']>-25)& (localityFeatures['LONG']<40)]
```

```python
all_occurrences = []
first_occurrences = []
last_occurrences = []
for timeUnit in timeUnits:
    all_occurrences.append(localityFeatures.loc[localityFeatures['TIMEUNIT']==timeUnit, \
                                                'NUMBER OF OCCURRENCES'].sum())
    first_occurrences.append(localityFeatures.loc[localityFeatures['TIMEUNIT']==timeUnit, \
                                                  'NUMBER OF FIRST OCCURRENCES'].sum())
    last_occurrences.append(localityFeatures.loc[localityFeatures['TIMEUNIT']==timeUnit, \
                                                  'NUMBER OF LAST OCCURRENCES'].sum())

all_occurrences = np.array(all_occurrences)
first_occurrences = np.array(first_occurrences)
last_occurrences = np.array(last_occurrences)
display(all_occurrences)
display(first_occurrences)
display(last_occurrences)
```

```python
label_fontsize = 20
tick_fontsize = 14
legend_fontsize = 16

fig, ax = plt.subplots(figsize=(20,5))
plt.plot(timeUnits, first_occurrences / all_occurrences)
plt.plot(timeUnits, last_occurrences / all_occurrences, linestyle='dashed')
plt.scatter(timeUnits,first_occurrences / all_occurrences)
plt.scatter(timeUnits,last_occurrences / all_occurrences)
#plt.title('PROPORTION OF FIRST/LAST OCCURRENCES')
plt.xlabel('TIME', fontsize=label_fontsize)
plt.ylabel('PROPORTION', fontsize=label_fontsize)
plt.legend(('first','last'), loc=(0.92,0.6), fontsize=legend_fontsize)
ax.tick_params(labelsize=tick_fontsize)
filepath = '../figures/proportion_first_occurrences.eps'
plt.savefig(filepath, bbox_inches='tight')
plt.show()

fig, ax = plt.subplots(figsize=(20,5))
plt.plot(timeUnits, all_occurrences)
plt.scatter(timeUnits,all_occurrences)
#plt.title('ALL OCCURRENCES')
plt.xlabel('TIME', fontsize=label_fontsize)
plt.ylabel('OCCURRENCES', fontsize=label_fontsize)
ax.tick_params(labelsize=tick_fontsize)
filepath = '../figures/all_occurrences.eps'
plt.savefig(filepath, bbox_inches='tight')
plt.show()
```

```python
localityFeatures['OCCURRENCES IN FOCAL AREA (t-1)'] = 0
localityFeatures['UNWEIGHTED OCCURRENCES IN FOCAL AREA (t-1)'] = 0
localityFeatures['OCCURRENCES IN FOCAL AREA (t+1)'] = 0
localityFeatures['UNWEIGHTED OCCURRENCES IN FOCAL AREA (t+1)'] = 0
localityFeatures['FIRST OCCURRENCES IN FOCAL AREA (t)'] = 0
localityFeatures['LAST OCCURRENCES IN FOCAL AREA (t)'] = 0
localityFeatures['OCCURRENCES IN FOCAL AREA (t)'] = 0
localityFeatures['UNWEIGHTED OCCURRENCES IN FOCAL AREA (t)'] = 0
localityFeatures['PROPORTION OF FIRST OCCURRENCES IN FOCAL AREA (t)'] = 0
localityFeatures['PROPORTION OF LAST OCCURRENCES IN FOCAL AREA (t)'] = 0
localityFeatures['HYP1'] = 0
localityFeatures['HYP2'] = 0
localityFeatures['HYP3'] = 0
localityFeatures['LOP0'] = 0
localityFeatures['LOP1'] = 0
localityFeatures['LOP2'] = 0

localityFeatures.head()
```

```python
# HYP and LOP data from these orders can be used for environmental estimation
permissibleOrders = ['Perissodactyla','Artiodactyla','Primates','Proboscidea','Hyracoidea','Dinocerata','Embrithopoda','Notoungulata','Astrapotheria','Pyrotheria','Litopterna']
print('NUMBER OF OCCURRENCES (includes "old" and "recent" time units)')
print()
print('Perissodactyla: ' + str(df[df['ORDER']=='Perissodactyla'].shape[0]))
print('Artiodactyla: ' + str(df[df['ORDER']=='Artiodactyla'].shape[0]))
print('Primates: ' + str(df[df['ORDER']=='Primates'].shape[0]))
print('Proboscidea: ' + str(df[df['ORDER']=='Proboscidea'].shape[0]))
print('Hyracoidea: ' + str(df[df['ORDER']=='Hyracoidea'].shape[0]))
print('Dinocerata: ' + str(df[df['ORDER']=='Dinocerata'].shape[0]))
print('Embrithopoda: ' + str(df[df['ORDER']=='Embrithopoda'].shape[0]))
print('Notoungulata: ' + str(df[df['ORDER']=='Notoungulata'].shape[0]))
print('Astrapotheria: ' + str(df[df['ORDER']=='Astrapotheria'].shape[0]))
print('Pyrotheria: ' + str(df[df['ORDER']=='Pyrotheria'].shape[0]))
print('Litopterna: ' + str(df[df['ORDER']=='Litopterna'].shape[0]))
```

```python
d = 5 # here we use a square around the focal locality with 2*d degree sides to find candidate localities
radius = 100 # radius (km) of the circle around the focal grid point; we look for localities within this circle

tic = time.clock()
count = 0 # to keep track of run progress
for ix in localityFeatures.index: 
    focalLONG = localityFeatures.loc[ix,'LONG'] 
    focalLAT = localityFeatures.loc[ix,'LAT'] 
    focalTimeUnit = localityFeatures.loc[ix,'TIMEUNIT'] 
    focalTimeUnitIndex = timeUnits.index(focalTimeUnit)

    # we look for candidate localities in the focal area one time unit before focalTimeUnit
    if focalTimeUnitIndex > 1:
        candidateLocalities = localityFeatures[(localityFeatures['TIMEUNIT']==timeUnits[focalTimeUnitIndex-1]) & (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
        for ix_candidate in candidateLocalities.index:
            distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], localityFeatures.loc[ix,['LAT','LONG']]).km
            if distance < radius:
                weight = (radius - distance)/radius
                localityFeatures.loc[ix,'OCCURRENCES IN FOCAL AREA (t-1)'] += weight * candidateLocalities.loc[ix_candidate,'NUMBER OF OCCURRENCES']
                localityFeatures.loc[ix,'UNWEIGHTED OCCURRENCES IN FOCAL AREA (t-1)'] += candidateLocalities.loc[ix_candidate,'NUMBER OF OCCURRENCES']

    # we look for candidate localities in the focal area one time unit after focalTimeUnit
    if focalTimeUnitIndex < 18:
        candidateLocalities = localityFeatures[(localityFeatures['TIMEUNIT']==timeUnits[focalTimeUnitIndex+1]) & (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
        for ix_candidate in candidateLocalities.index:
            distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], localityFeatures.loc[ix,['LAT','LONG']]).km
            if distance < radius:
                weight = (radius - distance)/radius
                localityFeatures.loc[ix,'OCCURRENCES IN FOCAL AREA (t+1)'] += weight * candidateLocalities.loc[ix_candidate,'NUMBER OF OCCURRENCES']
                localityFeatures.loc[ix,'UNWEIGHTED OCCURRENCES IN FOCAL AREA (t+1)'] += candidateLocalities.loc[ix_candidate,'NUMBER OF OCCURRENCES']

    # candidate localities in the current time unit
    candidateLocalities = localityFeatures[(localityFeatures['TIMEUNIT']==focalTimeUnit) & (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
    for ix_candidate in candidateLocalities.index:
        distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], localityFeatures.loc[ix,['LAT','LONG']]).km
        if distance < radius:
            weight = (radius - distance)/radius
            localityFeatures.loc[ix,'FIRST OCCURRENCES IN FOCAL AREA (t)'] += weight * candidateLocalities.loc[ix_candidate,'NUMBER OF FIRST OCCURRENCES']
            localityFeatures.loc[ix,'LAST OCCURRENCES IN FOCAL AREA (t)'] += weight * candidateLocalities.loc[ix_candidate,'NUMBER OF LAST OCCURRENCES']
            localityFeatures.loc[ix,'OCCURRENCES IN FOCAL AREA (t)'] += weight * candidateLocalities.loc[ix_candidate,'NUMBER OF OCCURRENCES']
            localityFeatures.loc[ix,'UNWEIGHTED OCCURRENCES IN FOCAL AREA (t)'] += candidateLocalities.loc[ix_candidate,'NUMBER OF OCCURRENCES']

            # environmental estimation
            occurrences = df.loc[(df['LIDNUM']==ix_candidate) & (df['ORDER'].isin(permissibleOrders))]
            for ix_occurrence in occurrences.index:
                if occurrences.loc[ix_occurrence,'HYP'] == 1:
                    localityFeatures.loc[ix,'HYP1'] += weight
                elif occurrences.loc[ix_occurrence,'HYP'] == 2:
                    localityFeatures.loc[ix,'HYP2'] += weight
                elif occurrences.loc[ix_occurrence,'HYP'] == 3:
                    localityFeatures.loc[ix,'HYP3'] += weight

                if occurrences.loc[ix_occurrence,'LOP'] == 0:
                    localityFeatures.loc[ix,'LOP0'] += weight
                elif occurrences.loc[ix_occurrence,'LOP'] == 1:
                    localityFeatures.loc[ix,'LOP1'] += weight
                elif occurrences.loc[ix_occurrence,'LOP'] == 2:
                    localityFeatures.loc[ix,'LOP2'] += weight       
        
    # show run progress
    count += 1
    if count % 1000 == 0:
        toc = time.clock()
        print('Progress: ' + str(count) + '/' + str(localityFeatures.shape[0]))
        print('Time elapsed: ' + str((toc-tic)/60) + ' minutes.')
        
toc = time.clock() 
print('Progress: ' + str(localityFeatures.shape[0]) + '/' + str(localityFeatures.shape[0]))
print('Time elapsed: ' + str((toc-tic)/60) + ' minutes.')         
```

```python
# calculate additional stats

localityFeatures['PROPORTION OF FIRST OCCURRENCES IN FOCAL AREA (t)'] = localityFeatures['FIRST OCCURRENCES IN FOCAL AREA (t)'] / localityFeatures['OCCURRENCES IN FOCAL AREA (t)']
localityFeatures['PROPORTION OF LAST OCCURRENCES IN FOCAL AREA (t)'] = localityFeatures['LAST OCCURRENCES IN FOCAL AREA (t)'] / localityFeatures['OCCURRENCES IN FOCAL AREA (t)']
localityFeatures['MEAN HYP'] = (localityFeatures['HYP1']*1 + localityFeatures['HYP2']*2 + localityFeatures['HYP3']*3) / (localityFeatures['HYP1'] + localityFeatures['HYP2'] + localityFeatures['HYP3']) 
localityFeatures['MEAN LOP'] = (localityFeatures['LOP1']*1 + localityFeatures['LOP2']*2) / (localityFeatures['LOP0'] + localityFeatures['LOP1'] + localityFeatures['LOP2']) 
localityFeatures['MAT'] = 24.7 + 13.8*localityFeatures['MEAN HYP']-25.1*localityFeatures['MEAN LOP']
localityFeatures['MAP'] = 2727.7 - 411.9*localityFeatures['MEAN HYP']-859.7*localityFeatures['MEAN LOP']
localityFeatures['NPP'] = 2957.8 - 304.3*localityFeatures['MEAN HYP']-1043.7*localityFeatures['MEAN LOP']

localityFeatures.head()
```
