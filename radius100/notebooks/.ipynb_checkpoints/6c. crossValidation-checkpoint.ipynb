# k-fold cross-validation

```python
# choose k for the number of folds in the k-fold cross-validation

def k_fold_mask(k):
    fold_size = int( (localities.shape[0] - (localities.shape[0] % k)) / k )

    mask = np.ones((k,localities.shape[0]))
    count = 0
    for i in range(0,k-1):
        mask[i,count:count+fold_size] = 0
        count += fold_size
    mask[-1,count:] = 0
    mask = np.array(mask, dtype=bool)
    
    # print some info on the mask
    print('Number of folds: ' + str(k))
    print('First fold size: ' + str((mask[0]==False).sum()))
    print('Last fold size: ' + str((mask[-1]==False).sum()))
    return mask
```

```python
# Logistic regression functions

def sigmoid(c,x):
    return 1/(1 + np.exp(-(c[0]+c[1]*x)))
def sigmoid2(c,x):
    return 1/(1 + np.exp(-(c[0]+c[1]*x[0]+c[2]*x[1])))
```

```python
display(localities.head())
display(localities.shape)
```

```python
# shuffle locality data
localities = shuffle(localities)
display(localities.head())
display(localities.shape)
```

```python
k = 10
mask = k_fold_mask(k)
```

# Mean test MSE and RMSE for first occurrences

```python
testMSE = []

for i in range(0,k):
    
    # create regression data
    localitiesRegression = localities.iloc[mask[i]]
    localitiesRegression = localitiesRegression[(localitiesRegression['TIMEUNIT'].isin(timeUnits[2:19]))]
    regressionFirst = np.zeros((int(localitiesRegression['NUMBER OF OCCURRENCES'].sum()), 3))
    dataIndex = 0
    for locality in localitiesRegression.index:    
        numberOccurrences = int(localities.loc[locality,'NUMBER OF OCCURRENCES'])

        # update regressionFirst
        numberNewOccurrences = int(localities.loc[locality,'NUMBER OF FIRST OCCURRENCES'])
        newArray = np.zeros((numberOccurrences,3))        
        newArray[:,0] = localities.loc[locality,'OCCURRENCES IN FOCAL AREA (t-1)']
        newArray[:,1] = localities.loc[locality,'OCCURRENCES IN FOCAL AREA (t)']
        if numberNewOccurrences>0:
            newOcc = np.ones((numberNewOccurrences))
            newArray[:numberNewOccurrences,2] = newOcc
        regressionFirst[dataIndex:dataIndex+numberOccurrences,:] = newArray

        # increase index
        dataIndex += numberOccurrences
        
    # Perform regression 
    X = regressionFirst[:,0:2] 
    X = sm.add_constant(X)
    y = regressionFirst[:,2]
    model = sm.Logit(y, X)
    result = model.fit()
    coefficients = result.params
    
    # test accuracy
    indexes = localities.iloc[mask[i]==False].index
    localities.loc[indexes, 'crossvalidation: EXPECTED PROPORTION OF FIRST OCCURRENCES'] = \
            sigmoid2(coefficients, [localities.loc[indexes, 'OCCURRENCES IN FOCAL AREA (t-1)'], \
                                        localities.loc[indexes, 'OCCURRENCES IN FOCAL AREA (t)']])
    localities.loc[indexes, 'RESIDUAL'] = \
            localities.loc[indexes, 'PROPORTION OF FIRST OCCURRENCES IN FOCAL AREA (t)'] - \
            localities.loc[indexes, 'crossvalidation: EXPECTED PROPORTION OF FIRST OCCURRENCES']
    
    MSE = (localities.loc[indexes, 'RESIDUAL']**2).mean()
    testMSE.append(MSE)
    print('MSE = ' + str(MSE))
    print('Progress: ' + str(i+1) + '/' + str(k))
```

```python
display('Average test MSE: ' + str( np.mean(testMSE) ))
display('Average test RMSE: ' + str( np.mean(np.sqrt(testMSE)) ))
```

# Mean test MSE for last occurrences

```python
testMSE = []

for i in range(0,k):
    
    # create regression data
    localitiesRegression = localities.iloc[mask[i]]
    localitiesRegression = localitiesRegression[(localitiesRegression['TIMEUNIT'].isin(timeUnits[1:18]))]
    regressionLast = np.zeros((int(localitiesRegression['NUMBER OF OCCURRENCES'].sum()), 3))
    dataIndex = 0
    for locality in localitiesRegression.index:    
        numberOccurrences = int(localities.loc[locality,'NUMBER OF OCCURRENCES'])

        # update regressionLast
        numberLastOccurrences = int(localities.loc[locality,'NUMBER OF LAST OCCURRENCES'])
        newArray = np.zeros((numberOccurrences,3))        
        newArray[:,0] = localities.loc[locality,'OCCURRENCES IN FOCAL AREA (t+1)']
        newArray[:,1] = localities.loc[locality,'OCCURRENCES IN FOCAL AREA (t)']
        if numberLastOccurrences>0:
            newOcc = np.ones((numberLastOccurrences))
            newArray[:numberLastOccurrences,2] = newOcc
        regressionLast[dataIndex:dataIndex+numberOccurrences,:] = newArray

        # increase index
        dataIndex += numberOccurrences
        
    # Perform regression 
    X = regressionLast[:,0:2] 
    X = sm.add_constant(X)
    y = regressionLast[:,2]
    model = sm.Logit(y, X)
    result = model.fit()
    coefficients = result.params
    
    # test accuracy
    indexes = localities.iloc[mask[i]==False].index
    localities.loc[indexes, 'crossvalidation: EXPECTED PROPORTION OF LAST OCCURRENCES'] = \
            sigmoid2(coefficients, [localities.loc[indexes, 'OCCURRENCES IN FOCAL AREA (t+1)'], \
                                        localities.loc[indexes, 'OCCURRENCES IN FOCAL AREA (t)']])
    localities.loc[indexes, 'RESIDUAL'] = \
            localities.loc[indexes, 'PROPORTION OF LAST OCCURRENCES IN FOCAL AREA (t)'] - \
            localities.loc[indexes, 'crossvalidation: EXPECTED PROPORTION OF LAST OCCURRENCES']
    
    MSE = (localities.loc[indexes, 'RESIDUAL']**2).mean()
    testMSE.append(MSE)
    print('MSE = ' + str(MSE))
    print('Progress: ' + str(i+1) + '/' + str(k))
```

```python
display('Average test MSE: ' + str( np.mean(testMSE) ))
display('Average test RMSE: ' + str( np.mean(np.sqrt(testMSE)) ))
```
