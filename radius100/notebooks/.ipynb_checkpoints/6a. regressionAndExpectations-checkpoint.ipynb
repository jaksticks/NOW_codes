# LOGISTIC REGRESSION AND EXPECTED PROPORTION OF FIRST AND LAST OCCURRENCES

Here, we perform a logistic regression to find out how many first and last occurrences we should expect in a given locality based on the number of occurrences in the preceding and following time units within the focal area of the locality. 

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

localityFeatures = pd.read_csv('../csv/localityFeatures100.csv', index_col=0)
timeUnits = ['old','MN1','MN2','MN3','MN4','MN5','MN6','MN7-8','MN9','MN10','MN11','MN12','MN13','MN14','MN15','MN16','MN17','MQ18','MQ19','recent']

pd.set_option('max_columns', 100)
```

```python
# Logistic regression curve

def sigmoid(c,x):
    return 1/(1 + np.exp(-(c[0]+c[1]*x)))
def sigmoid2(c,x):
    return 1/(1 + np.exp(-(c[0]+c[1]*x[0]+c[2]*x[1])))
```

```python
# Create regression data for first occurrences

# we only use timeunits MN2 to MQ19

localities = localityFeatures[(localityFeatures['TIMEUNIT'].isin(timeUnits[2:19]))]
#pre-allocate
regressionFirst = np.zeros((int(localities['NUMBER OF OCCURRENCES'].sum()), 3))
dataIndex = 0
for locality in localities.index:    
    numberOccurrences = int(localityFeatures.loc[locality,'NUMBER OF OCCURRENCES'])
    
    # update regressionFirst
    numberNewOccurrences = int(localityFeatures.loc[locality,'NUMBER OF FIRST OCCURRENCES'])
    newArray = np.zeros((numberOccurrences,3))        
    newArray[:,0] = localityFeatures.loc[locality,'OCCURRENCES IN FOCAL AREA (t-1)']
    newArray[:,1] = localityFeatures.loc[locality,'OCCURRENCES IN FOCAL AREA (t)']
    if numberNewOccurrences>0:
        newOcc = np.ones((numberNewOccurrences))
        newArray[:numberNewOccurrences,2] = newOcc
    regressionFirst[dataIndex:dataIndex+numberOccurrences,:] = newArray
    
    # increase index
    dataIndex += numberOccurrences

display(regressionFirst.shape)    
display(regressionFirst[0:20])
```

```python
# Create regression data for last occurrences

# we only use timeunits MN1 to MQ18
localities = localityFeatures[(localityFeatures['TIMEUNIT'].isin(timeUnits[1:18]))]
regressionLast = np.zeros((int(localities['NUMBER OF OCCURRENCES'].sum()), 3))
dataIndex = 0
for locality in localities.index:    
    numberOccurrences = int(localityFeatures.loc[locality,'NUMBER OF OCCURRENCES'])
    
    # update regressionLast
    numberLastOccurrences = int(localityFeatures.loc[locality,'NUMBER OF LAST OCCURRENCES'])
    newArray = np.zeros((numberOccurrences,3))        
    newArray[:,0] = localityFeatures.loc[locality,'OCCURRENCES IN FOCAL AREA (t+1)']
    newArray[:,1] = localityFeatures.loc[locality,'OCCURRENCES IN FOCAL AREA (t)']
    if numberLastOccurrences>0:
        newOcc = np.ones((numberLastOccurrences))
        newArray[:numberLastOccurrences,2] = newOcc
    regressionLast[dataIndex:dataIndex+numberOccurrences,:] = newArray
    
    # increase index
    dataIndex += numberOccurrences

display(regressionLast.shape)
display(regressionLast[0:20])
```

```python
# Perform regression for first occurrences

X = regressionFirst[:,0:2]
X = sm.add_constant(X)
y = regressionFirst[:,2]
model = sm.Logit(y, X)
result = model.fit()
display(result.summary())

coefficientsFirst = result.params
print('Coefficients:')
print(coefficientsFirst)
print()
print('95%-Confidence interval:')
confidence95_first = result.conf_int(alpha=0.05)
print(confidence95_first)
print()
print('p-values:')
print(result.pvalues)
```

```python
# REDO regression for first occurrences without 'OCCURRENCES IN FOCAL AREA (t)'

X = regressionFirst[:,0]
X = sm.add_constant(X)
y = regressionFirst[:,2]
model = sm.Logit(y, X)
result = model.fit()
display(result.summary())

coefficientsFirst = result.params
print('Coefficients:')
print(coefficientsFirst)
print()
print('95%-Confidence interval:')
confidence95_first = result.conf_int(alpha=0.05)
print(confidence95_first)
print()
print('p-values:')
print(result.pvalues)
```

```python
localityFeatures['OCCURRENCES IN FOCAL AREA (t-1)'].max()
```

```python
x = np.linspace(0,720,721)
fig, ax = plt.subplots(figsize=(20,5))
plt.plot(x,sigmoid(coefficientsFirst,x), 'red', linewidth=2)
plt.plot(x,sigmoid(confidence95_first[:,0],x), 'orange', linewidth=2)
plt.plot(x,sigmoid(confidence95_first[:,1],x), 'orange', linewidth=2)
plt.scatter(localityFeatures['OCCURRENCES IN FOCAL AREA (t-1)'], \
            localityFeatures['PROPORTION OF FIRST OCCURRENCES IN FOCAL AREA (t)'], alpha=0.5)
plt.xlabel('OCCURRENCES IN FOCAL AREA (t-1)')
plt.ylabel('EXPECTED PROPORTION OF FIRST \n OCCURRENCES IN FOCAL AREA (t)');
plt.show()
```

```python
# Perform regression for last occurrences

X = regressionLast[:,0:2]
X = sm.add_constant(X)
y = regressionLast[:,2]
model = sm.Logit(y, X)
result = model.fit()
display(result.summary())

coefficientsLast = result.params
print('Coefficients:')
print(coefficientsLast)
print()
print('95%-Confidence interval:')
confidence95_last = result.conf_int(alpha=0.05)
print(confidence95_last)
print()
print('p-values:')
print(result.pvalues)
```

```python
x = np.linspace(0,800,801)
y = np.linspace(0,800,801)
X,Y = np.meshgrid(x,y)

fig, ax = plt.subplots(figsize=(16,8))
CS = plt.contour(X,Y,sigmoid2(coefficientsLast,[X,Y]))
plt.clabel(CS, inline=1, fontsize=10)
color = localityFeatures['PROPORTION OF LAST OCCURRENCES IN FOCAL AREA (t)']
plt.scatter(localityFeatures['OCCURRENCES IN FOCAL AREA (t+1)'], \
            localityFeatures['OCCURRENCES IN FOCAL AREA (t)'], c=color, cmap='RdYlBu', alpha=0.4)
plt.colorbar()
plt.xlabel('OCCURRENCES IN FOCAL AREA (t+1)')
plt.ylabel('OCCURRENCES IN FOCAL AREA (t)')
plt.show()
```

```python
x = np.linspace(0,800,801)
y = 200
fig, ax = plt.subplots(figsize=(20,5))
plt.plot(x,sigmoid2(coefficientsLast,[x,y]), 'red', linewidth=2)
plt.plot(x,sigmoid2(confidence95_last[:,0],[x,y]), 'orange', linewidth=2)
plt.plot(x,sigmoid2(confidence95_last[:,1],[x,y]), 'orange', linewidth=2)
plt.scatter(localityFeatures['OCCURRENCES IN FOCAL AREA (t-1)'], \
            localityFeatures['PROPORTION OF LAST OCCURRENCES IN FOCAL AREA (t)'], alpha=0.5)
plt.title('REGRESSION ASSUMING OCCURRENCES (t) = ' + str(y))
plt.xlabel('OCCURRENCES IN FOCAL AREA (t-1)')
plt.ylabel('EXPECTED PROPORTION OF LAST \n OCCURRENCES IN FOCAL AREA (t)');
plt.show()
```

```python
x = np.linspace(0,800,801)
fig, ax = plt.subplots(figsize=(20,5))
plt.plot(x,sigmoid2(coefficientsLast,[x,0]), 'red', linewidth=2)
plt.plot(x,sigmoid2(coefficientsLast,[x,200]), 'orange', linewidth=2)
plt.plot(x,sigmoid2(coefficientsLast,[x,400]), 'yellow', linewidth=2)
plt.plot(x,sigmoid2(coefficientsLast,[x,600]), 'lightblue', linewidth=2)
plt.legend((0,200,400,600))
plt.scatter(localityFeatures['OCCURRENCES IN FOCAL AREA (t-1)'], \
            localityFeatures['PROPORTION OF LAST OCCURRENCES IN FOCAL AREA (t)'], alpha=0.5)
plt.title('REGRESSION ASSUMING DIFFERENT FIXED OCCURRENCES (t)')
plt.xlabel('OCCURRENCES IN FOCAL AREA (t-1)')
plt.ylabel('EXPECTED PROPORTION OF LAST \n OCCURRENCES IN FOCAL AREA (t)');
plt.show()
```

We can use the regressions to get an estimation for the exptected proportion of first and last occurrences.

```python
localityFeatures['EXPECTED PROPORTION OF FIRST OCCURRENCES'] = \
            sigmoid(coefficientsFirst, localityFeatures['OCCURRENCES IN FOCAL AREA (t-1)'])
localityFeatures['EXPECTED PROPORTION OF LAST OCCURRENCES'] = \
            sigmoid2(coefficientsLast, [localityFeatures['OCCURRENCES IN FOCAL AREA (t+1)'], \
                                        localityFeatures['OCCURRENCES IN FOCAL AREA (t)']])
```

```python
localityFeatures.to_csv('../csv/localityFeatures100.csv')
```

```python
# import grid data

gridFeaturesAll = []
for timeUnit in timeUnits[1:19]:
    filename = '../csv/gridFeatures100/gridFeatures_' + timeUnit + '.csv'
    gridFeatures = pd.read_csv(filename, index_col=0)
    gridFeaturesAll.append(gridFeatures)
```

```python
# use locality regressions to calculate expectations for the whole grid
for gridFeatures in gridFeaturesAll:
    gridFeatures['EXPECTED PROPORTION OF FIRST OCCURRENCES'] = \
            sigmoid(coefficientsFirst, gridFeatures['OCCURRENCES IN FOCAL AREA (t-1)'])
    gridFeatures['EXPECTED PROPORTION OF LAST OCCURRENCES'] = \
            sigmoid2(coefficientsLast, [gridFeatures['OCCURRENCES IN FOCAL AREA (t+1)'], \
                                        gridFeatures['OCCURRENCES IN FOCAL AREA (t)']])
```

```python
# save grid data

timeUnit_count = 1
for gridFeatures in gridFeaturesAll:
    focalTimeUnit = timeUnits[timeUnit_count]
    filename = '../csv/gridFeatures100/gridFeatures_' + focalTimeUnit + '.csv'
    gridFeatures.to_csv(filename)
    timeUnit_count += 1
```
