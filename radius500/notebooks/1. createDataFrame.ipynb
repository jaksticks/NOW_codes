# NOW DataFrame

Here we take the NOW raw data (.txt file) and turn it into a pandas DataFrame.

```python
import pandas as pd
pd.set_option('max_columns',100)
```

```python
with open('../NOW_rawdata/raw_data.txt','r') as file:    
    rawData = file.read()  
```

```python
rows = rawData.split('\n')
```

```python
finalData = []
for row in rows:
    splitRow = row.split(',')
    finalData.append(splitRow)
    
print(finalData[0:3])
```

```python
df = pd.DataFrame(finalData[1:],columns=finalData[0])
df.head()
```

```python
df.to_csv('../csv/NOW_dataframe.csv')
```
