# EMIGRATION AND IMMIGRATION

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
import time
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

df = pd.read_csv('../csv/NOW_dataframe_cleaned.csv', index_col=0)
localityFeatures = pd.read_csv('../csv/localityFeatures500.csv', index_col=0)
speciesFirstOccurrence = pd.read_csv('../csv/speciesFirstOccurrence.csv', index_col=0)
speciesLastOccurrence = pd.read_csv('../csv/speciesLastOccurrence.csv', index_col=0)

timeUnits = ['old','MN1','MN2','MN3','MN4','MN5','MN6','MN7-8','MN9','MN10','MN11','MN12','MN13','MN14','MN15','MN16','MN17','MQ18','MQ19','recent']
pd.set_option('max_columns',100)
```

```python
def sigmoid1(c,x):
    return 1/(1 + np.exp(-(c[0]+c[1]*x)))

def sigmoid2(c,x):
    return 1/(1 + np.exp(-(c[0]+c[1]*x[0]+c[2]*x[1])))
```

```python
d = 10 # here we use a square around the focal locality with 2*d degree sides to find candidate localities
radius = 500 # radius (km) of the circle around the focal grid point; we look for localities within this circle
emigrationRegressionData = np.array([]).reshape((0,3))
immigrationRegressionData = np.array([]).reshape((0,3))

tic = time.clock()
count = 0 # to keep track of run progress
for ix in localityFeatures.index: 
    focalTimeUnit = localityFeatures.loc[ix,'TIMEUNIT'] 
    focalTimeUnitIndex = timeUnits.index(focalTimeUnit)
    
    if (focalTimeUnitIndex > 0) & (focalTimeUnitIndex < 19):
        focalLONG = localityFeatures.loc[ix,'LONG'] 
        focalLAT = localityFeatures.loc[ix,'LAT'] 
    
        # we look for localities in the focal area in focalTimeUnit
        localities = localityFeatures[(localityFeatures['TIMEUNIT']==timeUnits[focalTimeUnitIndex]) & \
                      (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & \
                      (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
        for ix_candidate in localities.index:
            distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], \
                                localityFeatures.loc[ix,['LAT','LONG']]).km
            if distance >= radius:
                localities = localities.drop([ix_candidate])
            else:
                weight = (radius - distance)/radius
                # auxiliary info for later
                df.loc[df['LIDNUM']==ix_candidate, 'WEIGHT'] = weight 
        occurrencesNow = df[df['LIDNUM'].isin(localities.index)]
        speciesNow = set(occurrencesNow['ID'].unique())
        
        # species that originated at this time unit
        speciation = set(speciesFirstOccurrence[speciesFirstOccurrence[focalTimeUnit]==1].index)
        # remove species that only just originated
        speciesNow_immigration = speciesNow.difference(speciation)
        occurrencesNow_immigration = occurrencesNow.loc[occurrencesNow['ID'].isin(speciesNow_immigration)]
        # weighted number of current occurrences for species that exist in both current and previous time unit
        numberOccurrencesNow_immigration = occurrencesNow_immigration['WEIGHT'].sum()
        
        # species that did not go extinct at this time unit
        extinction = set(speciesLastOccurrence[speciesLastOccurrence[timeUnits[focalTimeUnitIndex]]==1].index)
        # remove species that went extinct
        speciesNow_emigration = speciesNow.difference(extinction) 
        occurrencesNow_emigration = occurrencesNow.loc[occurrencesNow['ID'].isin(speciesNow_emigration)]
        # weighted number of occurrences for species that exist in both current and next time unit
        numberOccurrencesNow_emigration = occurrencesNow_emigration['WEIGHT'].sum()
        # for focal locality, identify species that are immigrant and residents
        localitySpecies = set(df.loc[(df['LIDNUM']==ix) & df['ID'].isin(speciesNow_immigration), 'ID'])        
        localityFeatures.loc[ix, 'NUMBER IMMIGRANT AND RESIDENT'] = len(localitySpecies)
        # for focal locality, identify species that are emigrant and residents
        localitySpecies = set(df.loc[(df['LIDNUM']==ix) & df['ID'].isin(speciesNow_emigration), 'ID'])        
        localityFeatures.loc[ix, 'NUMBER EMIGRANT AND RESIDENT'] = len(localitySpecies)

        # IMMIGRATION
        if focalTimeUnitIndex > 1:
            # we look for localities in the focal area one time unit before focalTimeUnit
            localities = localityFeatures[(localityFeatures['TIMEUNIT']==timeUnits[focalTimeUnitIndex-1]) & \
                          (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & \
                          (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
            for ix_candidate in localities.index:
                distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], \
                                    localityFeatures.loc[ix,['LAT','LONG']]).km
                if distance >= radius:
                    localities = localities.drop([ix_candidate])
                else:
                    weight = (radius - distance)/radius
                    # auxiliary info for later
                    df.loc[df['LIDNUM']==ix_candidate, 'WEIGHT'] = weight 
            occurrencesPreviously = df[df['LIDNUM'].isin(localities.index)]
            speciesPreviously = set(occurrencesPreviously['ID'].unique())
            extinction = set(speciesLastOccurrence[speciesLastOccurrence[timeUnits[focalTimeUnitIndex-1]]==1].index)
            # species that exist in both current and previous time unit            
            speciesPreviously = speciesPreviously.difference(extinction)
            # remove occurrences for species going extinct
            occurrencesPreviously = occurrencesPreviously.loc[occurrencesPreviously['ID'].isin(speciesPreviously)]
            # weighted number of previous occurrences for species that exist in both current and previous time unit            
            numberOccurrencesPreviously = occurrencesPreviously['WEIGHT'].sum()
            # weighted number of immigrants now
            immigrantSpecies = speciesNow_immigration.difference(speciesPreviously)
            numberImmigrants = occurrencesNow_immigration.loc[occurrencesNow_immigration['ID'].\
                                              isin(immigrantSpecies), 'WEIGHT'].sum()
            # for focal locality, identify species that are immigrants
            localitySpecies = set(df.loc[(df['LIDNUM']==ix) & df['ID'].isin(speciesNow_immigration), 'ID'])
            localityImmigrantSpecies = localitySpecies.difference(speciesPreviously)
            localityNumberImmigrants = len(localityImmigrantSpecies)
            # add info to localityFeatures
            localityFeatures.loc[ix,'OCCURRENCES IMMIGRATION (t-1)'] = numberOccurrencesPreviously
            localityFeatures.loc[ix,'OCCURRENCES IMMIGRATION (t)'] = numberOccurrencesNow_immigration
            localityFeatures.loc[ix,'NUMBER IMMIGRANTS'] = numberImmigrants
            localityFeatures.loc[ix,'NUMBER IMMIGRANTS (LOCALITY)'] = localityNumberImmigrants
            
            # append immigrationRegressionData
            immigration1 = np.array([[numberOccurrencesPreviously, numberOccurrencesNow_immigration, 1]])            
            immigration1 = np.repeat(immigration1, localityNumberImmigrants, axis=0)
            immigration0 = np.array([[numberOccurrencesPreviously, numberOccurrencesNow_immigration, 0]])
            immigration0 = np.repeat(immigration0, localityFeatures.loc[ix, 'NUMBER IMMIGRANT AND RESIDENT']-\
                                     localityNumberImmigrants, axis=0)
            immigration = np.append(immigration1, immigration0, axis=0)
            immigrationRegressionData = np.append(immigrationRegressionData, immigration, axis=0)
        
        # EMIGRATION
        if focalTimeUnitIndex < 18:
            # we look for localities in the focal area one time unit after focalTimeUnit
            localities = localityFeatures[(localityFeatures['TIMEUNIT']==timeUnits[focalTimeUnitIndex+1]) & \
                          (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & \
                          (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
            for ix_candidate in localities.index:
                distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], \
                                    localityFeatures.loc[ix,['LAT','LONG']]).km
                if distance >= radius:
                    localities = localities.drop([ix_candidate])
                else:
                    weight = (radius - distance)/radius
                    # auxiliary info for later
                    df.loc[df['LIDNUM']==ix_candidate, 'WEIGHT'] = weight 
            occurrencesNext = df[df['LIDNUM'].isin(localities.index)]
            speciesNext = set(occurrencesNext['ID'].unique())
            speciation = set(speciesFirstOccurrence[speciesFirstOccurrence[timeUnits[focalTimeUnitIndex+1]]==1].index)
            # species that exist in both current and next time unit            
            speciesNext = speciesNext.difference(speciation)
            # remove occurrences for species that originated in the next time unit
            occurrencesNext = occurrencesNext.loc[occurrencesNext['ID'].isin(speciesNext)]
            # weighted number of next occurrences for species that exist in both current and next time unit            
            numberOccurrencesNext = occurrencesNext['WEIGHT'].sum()
            # weighted number of emigrants now
            emigrantSpecies = speciesNow_emigration.difference(speciesNext)
            numberEmigrants = occurrencesNow_emigration.loc[occurrencesNow_emigration['ID'].\
                                              isin(emigrantSpecies), 'WEIGHT'].sum()
            # for focal locality, identify species that are emigrants
            localitySpecies = set(df.loc[(df['LIDNUM']==ix) & df['ID'].isin(speciesNow_emigration), 'ID'])
            localityEmigrantSpecies = localitySpecies.difference(speciesNext)
            localityNumberEmigrants = len(localityEmigrantSpecies)
            # add info to localityFeatures
            localityFeatures.loc[ix,'OCCURRENCES EMIGRATION (t+1)'] = numberOccurrencesNext
            localityFeatures.loc[ix,'OCCURRENCES EMIGRATION (t)'] = numberOccurrencesNow_emigration
            localityFeatures.loc[ix,'NUMBER EMIGRANTS'] = numberEmigrants
            localityFeatures.loc[ix,'NUMBER EMIGRANTS (LOCALITY)'] = localityNumberEmigrants
            
            # append emigrationRegressionData
            emigration1 = np.array([[numberOccurrencesNext, numberOccurrencesNow_emigration, 1]])            
            emigration1 = np.repeat(emigration1, localityNumberEmigrants, axis=0)
            emigration0 = np.array([[numberOccurrencesNext, numberOccurrencesNow_emigration, 0]])
            emigration0 = np.repeat(emigration0, localityFeatures.loc[ix, 'NUMBER EMIGRANT AND RESIDENT']-\
                                    localityNumberEmigrants, axis=0)
            emigration = np.append(emigration1, emigration0, axis=0)
            emigrationRegressionData = np.append(emigrationRegressionData, emigration, axis=0)            
        
    # show run progress
    count += 1
    if count % 1000 == 0:
        toc = time.clock()
        print('Progress: ' + str(count) + '/' + str(localityFeatures.shape[0]))
        print('Time elapsed: ' + str((toc-tic)/60) + ' minutes.')
        
toc = time.clock() 
print('Progress: ' + str(localityFeatures.shape[0]) + '/' + str(localityFeatures.shape[0]))
print('Time elapsed: ' + str((toc-tic)/60) + ' minutes.')  
```

```python
# regression emigration
X = emigrationRegressionData[:,0:2]
X = sm.add_constant(X)
y = emigrationRegressionData[:,2]
model = sm.Logit(y, X)
result = model.fit()
display(result.summary())
coefficients_e = result.params
print('Coefficients:')
print(coefficients_e)
```

```python
# regression immigration
X = immigrationRegressionData[:,0:2]
X = sm.add_constant(X)
y = immigrationRegressionData[:,2]
model = sm.Logit(y, X)
result = model.fit()
display(result.summary())
coefficients_i = result.params
print('Coefficients:')
print(coefficients_i)
```

## gridFeatures

Gather necessary information for each grid cell.

```python
# import grid data

gridFeaturesAll = []
for timeUnit in timeUnits[1:19]:
    filename = '../csv/gridFeatures500/gridFeatures_' + timeUnit + '.csv'
    gridFeatures = pd.read_csv(filename, index_col=0)
    gridFeaturesAll.append(gridFeatures)
```

```python
d = 10 # here we use a square around the focal locality with 2*d degree sides to find candidate localities
radius = 500 # radius (km) of the circle around the focal grid point; we look for localities within this circle

timeUnitCount = 1
for gridFeatures in gridFeaturesAll:
    tic = time.clock()
    count = 0 # to keep track of run progress    
    for ix in gridFeatures.index: 
        focalLONG = gridFeatures.loc[ix,'LONG'] 
        focalLAT = gridFeatures.loc[ix,'LAT'] 
        focalTimeUnitIndex = timeUnitCount
        focalTimeUnit = timeUnits[focalTimeUnitIndex]
        
        # we look for localities in the focal area in focalTimeUnit
        localities = localityFeatures[(localityFeatures['TIMEUNIT']==timeUnits[focalTimeUnitIndex]) & \
                      (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & \
                      (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
        for ix_candidate in localities.index:
            distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], \
                                gridFeatures.loc[ix,['LAT','LONG']]).km
            if distance >= radius:
                localities = localities.drop([ix_candidate])
            else:
                weight = (radius - distance)/radius
                # auxiliary info for later
                df.loc[df['LIDNUM']==ix_candidate, 'WEIGHT'] = weight 
        occurrencesNow = df[df['LIDNUM'].isin(localities.index)]
        speciesNow = set(occurrencesNow['ID'].unique())

        # species list without species that originated in this time unit
        speciation = set(speciesFirstOccurrence[speciesFirstOccurrence[focalTimeUnit]==1].index)        
        speciesNowNoSpeciation = speciesNow.difference(speciation)
        occurrencesNowNoSpeciation = occurrencesNow.loc[occurrencesNow['ID'].isin(speciesNowNoSpeciation)]
        # weighted occurrences now disregarding species that just originated
        gridFeatures.loc[ix,'OCCURRENCES WITHOUT SPECIATION'] = occurrencesNowNoSpeciation['WEIGHT'].sum()

        # species list without species that went extinct
        extinction = set(speciesLastOccurrence[speciesLastOccurrence[focalTimeUnit]==1].index)
        speciesNowNoExtinction = speciesNow.difference(extinction)
        # weighted occurrences now disregarding species that are going extinct
        gridFeatures.loc[ix,'OCCURRENCES WITHOUT EXTINCTION'] = \
                occurrencesNow.loc[occurrencesNow['ID'].isin(speciesNowNoExtinction), 'WEIGHT'].sum()
        
        if timeUnitCount>1:
            # we look for localities in the focal area in focalTimeUnit-1
            localities = localityFeatures[(localityFeatures['TIMEUNIT']==timeUnits[focalTimeUnitIndex-1]) & \
                          (localityFeatures['LONG']>focalLONG-d) & (localityFeatures['LONG']<focalLONG+d) & \
                          (localityFeatures['LAT']>focalLAT-d) & (localityFeatures['LAT']<focalLAT+d)]
            for ix_candidate in localities.index:
                distance = geodesic(localityFeatures.loc[ix_candidate,['LAT','LONG']], \
                                    gridFeatures.loc[ix,['LAT','LONG']]).km
                if distance >= radius:
                    localities = localities.drop([ix_candidate])
                else:
                    weight = (radius - distance)/radius
                    # auxiliary info for later
                    df.loc[df['LIDNUM']==ix_candidate, 'WEIGHT'] = weight 
            occurrencesPreviously = df[df['LIDNUM'].isin(localities.index)]
            speciesPreviously = set(occurrencesPreviously['ID'].unique())

            # weighted occurrences of immigrated species 
            gridFeatures.loc[ix,'IMMIGRATION'] = occurrencesNowNoSpeciation.loc\
                                    [~occurrencesNowNoSpeciation['ID'].isin(speciesPreviously), 'WEIGHT'].sum()
            
            # species list without species that went extinct
            extinction = set(speciesLastOccurrence[speciesLastOccurrence[timeUnits[focalTimeUnitIndex-1]]==1].index)
            speciesPreviouslyNoExtinction = speciesPreviously.difference(extinction)
            # previous occurrences for species that did not go extinct
            occurrencesPreviouslyNoExtinction = occurrencesPreviously.loc[occurrencesPreviously['ID'].\
                                                                  isin(speciesPreviouslyNoExtinction)]            
            # weighted occurrences of emigrated species 
            gridFeatures.loc[ix,'EMIGRATION'] = occurrencesPreviouslyNoExtinction.loc\
                    [~occurrencesPreviouslyNoExtinction['ID'].isin(speciesNowNoSpeciation), 'WEIGHT'].sum()
        
        # show run progress
        count += 1
        if count % 5000 == 0:
            toc = time.clock()
            print('Progress: ' + str(count) + '/' + str(gridFeatures.shape[0]))
            print('Time elapsed: ' + str((toc-tic)/60) + ' minutes.')

    print(focalTimeUnit)
    timeUnitCount += 1
    toc = time.clock() 
    print('Progress: ' + str(gridFeatures.shape[0]) + '/' + str(gridFeatures.shape[0]))
    print('Time elapsed: ' + str((toc-tic)/60) + ' minutes.') 
    print() 
```

```python
# use locality regressions to calculate expectations for the whole grid
count=1
for gridFeatures in gridFeaturesAll[1::]:
    gridFeatures['EXPECTED PROPORTION OF EMIGRATION'] = sigmoid2(coefficients_e, \
        [gridFeatures['OCCURRENCES WITHOUT SPECIATION'], gridFeaturesAll[count-1]['OCCURRENCES WITHOUT EXTINCTION']])
    gridFeatures['EXPECTED PROPORTION OF IMMIGRATION'] = sigmoid2(coefficients_i, \
        [gridFeaturesAll[count-1]['OCCURRENCES WITHOUT EXTINCTION'], gridFeatures['OCCURRENCES WITHOUT SPECIATION']])
    count += 1
```
